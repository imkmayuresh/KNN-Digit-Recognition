```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfx = pd.read_csv('xdata.csv')
dfy = pd.read_csv('ydata.csv')
```

```python
X = dfx.values
Y = dfy.values

X = X[:,1:]
Y = Y[:,1:].reshape((-1,))

print(X)
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
query_x = np.array([2,3])
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(query_x[0],query_x[1],color='blue')
plt.show()
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,queryPoint,k=5):
    
    vals=[]
    
    m = X.shape[0]
    
    for i in range(m):
        d = dist(queryPoint,X[i])
        vals.append((d,Y[i]))
        
    vals=sorted(vals)
    vals = vals[:k]
        
    vals = np.array(vals)
    
    new_vals=np.unique(vals[:,1],return_counts=True)
    print(new_vals)
    
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    
    return pred
```

```python
knn(X,Y,query_x)
```
